### **1.5	OPTIONAL Problem - Python + Google**

Python has a function called zip() that might be helpful in solving the issue mentioned in Problem 3.
```python
station_names = ['Braemar', 'Dunstaffnage', 'Eskdalemuir', 'Leuchars', 'Nairn', 'Newton Rigg', 'Paisley', 'Wick Airport']
station_start_years = [1959, 1972, 1914, 1957, 1931, 1959, 1959, 1914]
station_names.extend(new_stations)
station_start_years.extend(new_years)
print(station_names)
print(station_start_years)
met_station = zip(station_names, station_start_years)
zip_met_station = list(met_station)
met_station_sort = sorted(zip_met_station,key =lambda x:x[1] )
print(met_station_sort)
# lambda is a anonymous standalone function, i.e. it isn't defined as part of the code and it operates within a single line.
# x:x is the whole zipped list of tuples. If it contained more than two elements in each tupple then it is still called as x:x
# x:x[0] is the first element of the tuple
# x:x[1] is the second element of the tuple
# x:x[2] would be the third if there were 3 elements stored in the tupples such as lat long.
```
